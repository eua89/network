```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../gpatounas.csv")
```

```python
indx = [i for i in range(df.shape[0])]
```

```python
# df['eCell_calc'] = df[['eCell','Calculation']].apply(lambda x: "_".join(x), axis=1 )
# df = df.set_index('eCell_calc')
df = df.drop('Unnamed: 0', 1)

df_calc = df.drop('eCell', 1)
df_calc = df_calc.set_index([indx,'Calculation'])

df_ecell = df.drop('Calculation', 1)
df_ecell = df_ecell.set_index([indx,'eCell'])
```

```python
df_ecell.index[0]
```

```python
df_calc_trans= df_calc.T
df_calc_trans.shape

df_ecell_trans = df_ecell.T
df_ecell_trans.shape
```

```python
df_ecell_trans.index
```

```python
df_ecell_trans.index = pd.to_datetime(df_ecell_trans.index)
df_ecell_trans.head()

df_calc_trans.index = pd.to_datetime(df_calc_trans.index)
df_calc_trans.head()
```

```python
df_ecell_trans = df_ecell_trans.apply(lambda x: x.str.replace(",","."))
df_ecell_trans = df_ecell_trans.apply(lambda x: x.str.replace("%",""))

df_calc_trans = df_calc_trans.apply(lambda x: x.str.replace(",","."))
df_calc_trans = df_calc_trans.apply(lambda x: x.str.replace("%",""))
```

```python
df_ecell_trans = df_ecell_trans.apply(pd.to_numeric)

df_calc_trans = df_calc_trans.apply(pd.to_numeric)
```

```python
df_ecel_no_zero = df_ecell_trans.loc[:, (df_ecell_trans != 0).any(axis=0)]
df_ecel_no_zero.shape

df_calc_no_zero = df_calc_trans.loc[:, (df_calc_trans != 0).any(axis=0)]
df_calc_no_zero.shape
```

```python
df_ecel_no_zero.shape
```

# Delete all calcs or ecells when one is zero

```python
empty_ecell =[]
for enum,i in enumerate((df_ecell_trans != 0).any(axis=0)):
    if i==False:
        empty_ecell.append(df_ecell_trans.columns[enum][1])
        empty_ecell=list(set(empty_ecell))
        
empty_calc =[]
for enum,i in enumerate((df_calc_trans != 0).any(axis=0)):
    if i==False:
        empty_calc.append(df_calc_trans.columns[enum][1])
        empty_calc=list(set(empty_calc))
        
print(len(empty_ecell))
print(len(empty_calc))
```

```python
df_drop_all_zeros_ecell = df_ecell_trans
for col in empty_ecell:
#     df_drop_all_zeros_ecell = df_ecell_trans[df_ecell_trans.columns.drop(list(df_ecell_trans.filter(regex=col)))]
    cols_to_drop = [c for c in df_drop_all_zeros_ecell.columns if col in c]
    df_drop_all_zeros_ecell= df_drop_all_zeros_ecell.drop(cols_to_drop, axis=1)
```

```python
calc_uniq=[]
for i in range(len(df_calc_no_zero.columns)):
    calc_uniq.append(df_calc_no_zero.columns[i][1])
    
ecell_uniq=[]
for i in range(len(df_ecel_no_zero.columns)):
    ecell_uniq.append(df_ecel_no_zero.columns[i][1])
```

```python
calc_uniq = list(set(calc_uniq))

ecell_uniq = list(set(ecell_uniq))
```

```python
calc_corr = []
for i in calc_uniq:
    calc_corr.append(df_calc_no_zero.filter(regex=i).corr(method='pearson'))
    
ecell_corr = []
for i in ecell_uniq:
    ecell_corr.append(df_ecel_no_zero.filter(regex=i).corr(method='pearson'))
```
