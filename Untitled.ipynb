```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("gpatounas.csv")
```

```python
df = df.set_index('Calculation')
df.head()
```

```python
df = df.drop('Unnamed: 0', 1)
df = df.drop('eCell', 1)
df.shape
```

```python
df= df.T
df.shape
```

```python
df.index = pd.to_datetime(df.index)
df.head()
```

```python
df.apply(lambda x: x.str.replace(',','.'))
lst = list(df.columns)
df[lst]=df[lst].astype(float)
```

```python
df = df.replace(0,np.nan).dropna(axis=1,how="all")
```
