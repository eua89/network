```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../gpatounas.csv")
```

```python
df.head()
```

```python
df['eCell_calc'] = df[['eCell','Calculation']].apply(lambda x: "_".join(x), axis=1 )
df = df.set_index('eCell_calc')
df = df.drop('Unnamed: 0', 1)
df = df.drop('eCell', 1)
df = df.drop('Calculation', 1)
df.head()
```

```python
df.shape
```

```python
df_trans= df.T
df_trans.shape
```

```python
# df_trans.dtypes
```

```python
df_trans.index = pd.to_datetime(df_trans.index)
df_trans.head()
```

```python
df_trans = df_trans.apply(lambda x: x.str.replace(",","."))
df_trans = df_trans.apply(lambda x: x.str.replace("%",""))
# lst = list(df.columns)
# df[lst]=df[lst].astype(float)
```

```python
df_trans = df_trans.apply(pd.to_numeric)
```

```python
df_no_zero = df_trans.loc[:, (df_trans != 0).any(axis=0)]
df_no_zero.shape
```

```python
corr = df_no_zero.corr(method='pearson')
```

```python
corr
```

```python
for col in corr.columns:
    pos_rows = corr.loc[corr[col]>0.8]
    print ('For the {} column the rows are {}'.format(col, pos_rows))
```
